```python
import time
import pandas as pd
import requests
```

```python
name = []
species = []
status = []
count = 0
for i in range(1,11):
    url = f"https://rickandmortyapi.com/api/character/{i}"
    resp = requests.get(url)
    result  = resp.json()
    name.append(result['name'])
    species.append(result['species'])
    status.append(result['status'])
    time.sleep(1) #seconds
    count += 1
    print(count)
#url = "https://swapi.dev/api/people/"
df = pd.DataFrame({
    "name":name,"species":species,"status":status
})
df
```

```python
# sklearn basics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# read data
test11 = pd.read_csv("https://gist.githubusercontent.com/Wizerlous/97826525a05111f3dc7629727c117802/raw/7ad3b8e90d9d55a94994faa48e50b50b14c89c7e/testfile.csv")
test11
```

```python
# prepare data
X = test11.drop(["ALL_EL", "PROV_CODE","PROV_NAME","N","NE","M","E","W","S","Years","region"], axis=1)
y = test11["ALL_EL"]

# split data
X_train, X_test, y_train, y_test = train_test_split(
	X, y, test_size = 0.25, random_state = 42 #set.seed()
)
```

```python
X_train.shape # ( row, column )
```

```python
X_train
```

```python
# train model
model = LinearRegression()
model.fit(X_train, y_train)

# test model
p = model.predict(X_test)
print(p)
```

```python
model.score(X_test, y_test) # R2
```
